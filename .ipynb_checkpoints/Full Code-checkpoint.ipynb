```python
import pandas as pd
import ast
import altair as alt
from vega_datasets import data
```

# LibreBOR insolvencias

## 1. Get Flat Data

```python
filename = "data/insolvencias_2021.04.20"
json_df = pd.read_json(filename + '.json', encoding='UTF-8')
df = json_df
```

```python
json_df
```

```python
df_i = df.insolvencias.apply(pd.Series)
df_i["Company"] = [ast.literal_eval(str(x).replace("[","").replace("]","")) for x in df_i.company]
df_i_company = df_i.join(df_i.Company.apply(pd.Series))
df_i_company

df_i_c = df_i_company
df_i_c["Announcements"] = [ast.literal_eval(str(x).replace("[","").replace("]","")) for x in df_i_c.announcements]
df_i_c.Announcements.apply(pd.Series)
df_full = df_i_c.join(df_i_c.Announcements.apply(pd.Series))
df_full
```

```python
df_full.to_csv(filename + '.csv')
```

## 2. Read Flat Data

```python
filename = "data/insolvencias_2021.04.20"

df = pd.read_csv(filename + '.csv', low_memory = False)
```

```python
df.shape
```

```python
df["day"]= [x[:7] for x in df["date"]]
df.sort_values(by = ["day"],inplace = True)
df.head(5)
```

```python
df.groupby(['day']).size()
```

```python
df_dates = pd.DataFrame(df.groupby(['day']).size()).reset_index("day")
df_dates
```

```python
df_dates_cn = df_dates.rename(columns = {0:"recuento_insolvencias", "day":"month"})
df_dates_cn
```

```python
chart = alt.Chart(df_dates_cn).mark_line().encode(
    x='month:T',
    y='recuento_insolvencias:Q'
).properties(
    width=800,
    height=300
)
chart.save('img/insolvencias_2021.04.20.html')

chart
```

```python

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)
    
plt.plot(df_dates.day, df_dates[0])
plt.title('Recuento de registro por fechas')
plt.xlabel('Date')
plt.ylabel('Count filas')
plt.show()
```
